```python
evadidos = evasao[evasao['SITUACAO'].str.lower() == 'evadido']
evadidos_por_ano = evadidos.groupby('ANO_INGRESSO').size()
evadidos_por_ano = evadidos_por_ano.sort_index()
plt.plot(evadidos_por_ano.index, evadidos_por_ano.values, marker='o', linestyle='-')
plt.xlabel('Ano')
plt.ylabel('Número de Evadidos')
plt.title('Linha do Tempo de Evadidos por Ano de ingresso')
plt.grid(True)
plt.show()
```
