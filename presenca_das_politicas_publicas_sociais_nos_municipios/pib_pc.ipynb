```python

import pandas as pd
import zipfile
```

```python

# Nome do arquivo zipado
zip_file_name = 'base_de_dados.zip'

# Nome do arquivo Excel dentro do zip
excel_file_name = 'PIB dos Municípios - base de dados 2010-2020.xls'

with zipfile.ZipFile(zip_file_name, 'r') as zip_ref:
    zip_ref.extract(excel_file_name)

# Carregue o arquivo Excel em um DataFrame
df = pd.read_excel(excel_file_name)

# Exiba o DataFrame
df.head()
```

```python
#vamos comerçar os gráficos. O primeiro será um ranking dos dez munícipios com menor IGD-M
!pip install matplotlib
import matplotlib.pyplot as plt

# Ordeando os valores de 'igd_m' em ordem decrescente para os 10 municipios com menor taxa
top_10_municipios_menor = df.sort_values(by='pib_pc', ascending=True).head(10)

```

```python
top_10_municipios_menor
```

```python
# Gráfico para os 10 municípios com menor pib

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 4))

# Plota o gráfico de barras horizontais no subplot 'ax'
ax.barh(top_10_municipios_menor['municipio'], top_10_municipios_menor['pib_pc'], color='#6B2D58')

# Configurações do subplot 'ax'
ax.set_xlabel('pib')
ax.set_ylabel('municipio')
ax.set_title('10 municípios com menor PIB per capta')
ax.tick_params(axis='x')  # Rotaciona os rótulos do eixo x

# Exibe a figura
plt.show()
```

```python
#Agora a mesma coisa para os dez municípios com as maiores pib
top_10_municipios_maior = df.sort_values(by='pib_pc', ascending=False).head(10)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 4))

# Plota o gráfico de barras horizontais no subplot 'ax'
ax.barh(top_10_municipios_maior['municipio'], top_10_municipios_maior['pib_pc'], color='#6B2D58')

# Configurações do subplot 'ax'
ax.set_xlabel('pib')
ax.set_ylabel('municipio')
ax.set_title('10 municípios com maior PIB per capta')
ax.tick_params(axis='x')  # Rotaciona os rótulos do eixo x

# Exibe a figura
plt.show()
```

```python
#dividindo o dataframe em 10 grupos iguais
Q1 = df['pib_pc'].quantile(0.1)
Q2 = df['pib_pc'].quantile(0.2)
Q3 = df['pib_pc'].quantile(0.3)
Q4 = df['pib_pc'].quantile(0.4)
Q5 = df['pib_pc'].quantile(0.5)  # Mediana (quintil 5)
Q6 = df['pib_pc'].quantile(0.6)
Q7 = df['pib_pc'].quantile(0.7)
Q8 = df['pib_pc'].quantile(0.8)
Q9 = df['pib_pc'].quantile(0.9)

print('Primeiro décimo: ', Q1)
print('Segundo décimo: ', Q2)
print('Terceiro décimo: ', Q3)
print('Quarto décimo: ', Q4)
print('Mediana (Quintil 5): ', Q5)
print('Sexto décimo: ', Q6)
print('Sétimo décimo: ', Q7)
print('Oitavo décimo: ', Q8)
print('Nono décimo: ', Q9)
```

```python
#nomemando os grupos e criando uma nova coluna para o rank deles
df['QuintilRank'] = pd.qcut(
    df['pib_pc'],
    q=10,
    labels=[
        "primeiro décimo",
        "segundo décimo",
        "terceiro décimo",
        "quarto décimo ",
        "quinto décimo ",
        "sexto décimo ",
        "sétimo décimo",
        "oitavo décimo",
        "nono décimo",
        "décimo décimo"
    ])
```

```python
df
```

```python
# verificando quantos municipios ficaram em cada grupo
contagem_quintis = df['QuintilRank'].value_counts()
```

```python
contagem_quintis
```

```python
# calculando a média de cada quintil com base no valor de pib_pc
media_quintil = df.groupby('QuintilRank')['pib_pc'].mean()
```

```python
media_quintil
```

```python
#Criando um gráfico de barras com essas médias
media_quintil = df.groupby('QuintilRank')['pib_pc'].mean()

# Plotar o gráfico de barras das médias por quintil
media_quintil.plot(kind='bar', figsize=(10, 6))
plt.title('Média do PIB per capta por Quintil')
plt.xlabel('Quintil')
plt.ylabel('Média do PIB per capta')
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x, se necessário
plt.show()
```

```python
#finalmente um histograma, para verificarmos a frequencia que os valores aparecem em cada quintil 
dados = df['pib_pc']

# Suponha que 'media_quintil' já contenha as médias dos quintis calculadas anteriormente.
# Transforme a série em uma lista de valores.
medias_quintis = media_quintil.values

# Plotar o histograma com as médias dos quintis como bins
plt.figure(figsize=(10, 6))  # Tamanho da figura
plt.hist(dados, bins=medias_quintis, color='skyblue', edgecolor='black')
plt.title('Histograma do PIB ')
plt.xlabel('PIB per capta')
plt.ylabel('Frequência')
plt.show()
```

```python
import matplotlib.pyplot as plt

# criando um gráfico de pizza para verificar a representação de cada quintil no todo
media_quintil = df.groupby('QuintilRank')['pib_pc'].mean()

# Criação do gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(media_quintil, labels=media_quintil.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribuição da média do PIB per capta por quintil')
plt.axis('equal')  # Garante que o gráfico de pizza seja circular
plt.show()
```
