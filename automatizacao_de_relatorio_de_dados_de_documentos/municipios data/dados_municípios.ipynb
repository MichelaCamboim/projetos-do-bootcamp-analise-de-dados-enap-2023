# Dados Populacionais do Censo 2022, por Municípios

```python
import pandas as pd
import re
import unicodedata

#A Daiana melhorou/reduziu bem esse código, mas vou manter aqui, por conta da limpeza dos nomes dos municípios

#Ler o arquivo em excel
df_mun_2022 = pd.read_excel('POP2022_Municipios_20230622.xls')

#Organizar os nomes das colunas e retirar linhas de informações, que contém campos nulos
df_mun_2022.columns = df_mun_2022.iloc[0]
df_mun_2022= df_mun_2022.iloc[1:].dropna()

#Limpar a coluna população, retirando observações de notas de rodapé e transformando-a em número inteiro
df_mun_2022['POPULAÇÃO'] = df_mun_2022['POPULAÇÃO'].astype(str)
df_mun_2022['POPULAÇÃO'] = df_mun_2022['POPULAÇÃO'].apply(lambda x: re.sub(r'\([^)]*\)', '', x))
df_mun_2022['POPULAÇÃO'] = df_mun_2022['POPULAÇÃO'].apply(lambda x: x.replace('.','')).astype(int)

#Criar a coluna de unidade geográfica, com o código completo de cada município
df_mun_2022['geocodigo'] = df_mun_2022['COD. UF'].astype(str) + df_mun_2022['COD. MUNIC'].astype(str) #usando o mesmo nome que a Daiana usou no 2010

#Retirar as colunas que não serão utilizadas
retirar = ['UF', 'COD. UF', 'COD. MUNIC']
df_mun_2022 = df_mun_2022.drop(retirar, axis=1)

#Limpando as strings dos nomes dos municípios
df_mun_2022['NOME DO MUNICÍPIO'] = df_mun_2022['NOME DO MUNICÍPIO'].apply(lambda x: unicodedata.normalize("NFD", x))
df_mun_2022['NOME DO MUNICÍPIO'] = df_mun_2022['NOME DO MUNICÍPIO'].apply(lambda x: x.encode("ascii", "ignore"))
df_mun_2022['NOME DO MUNICÍPIO'] = df_mun_2022['NOME DO MUNICÍPIO'].apply(lambda x: x.decode("utf-8"))
df_mun_2022['NOME DO MUNICÍPIO'] = df_mun_2022['NOME DO MUNICÍPIO'].apply(lambda x: x.lower())
df_mun_2022['NOME DO MUNICÍPIO'] = df_mun_2022['NOME DO MUNICÍPIO'].apply(lambda x:x.strip())

df_mun_2022.head()
```
